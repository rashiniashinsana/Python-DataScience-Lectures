###  Q. Fetch a webpage, check if the request was successful, and extract the main heading text from the page's content.You can assume that the base url is 'http://example.com'.

```python
import requests

def fetch_weather_data(latitude,longitude,current):
    base_url ='https://api.open-meteo.com/v1/forecast'

    params ={
        'latitude': latitude,
        'longitude': longitude,
        'current':current
    }

    response = requests.get(base_url,params=params)

    if(response.status_code == 200):
        data = response.json()
        return data
    else:
        return None
    
def parse_weather_data(data):
    temperature = data.get('temperature_2m')
    wind_speed = data.get('wind_speed_10m')
    
    print('Temp',temperature)
    print('Wind_speed',wind_speed)

data = fetch_weather_data(52.5074,-0.1278,'temperature_2m,wind_speed_10m')
if data:
    print(data)
    parse_weather_data(data.get('current'))
else:
    print('Failed to fetch data')


```
